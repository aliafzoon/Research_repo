```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
MODEL_ADDRESS = "./models/trained/v1.2/34000/final model.h5"
MINMAX_PATH = "./data/minmax all.xlsx"
TEST_FILE_ADDRESS = "TEST_FILE.XLSX"
```

```python
def unnormer(df, minmax):
    src = pd.read_excel(minmax, engine="openpyxl")
    minv = src.iloc[0].to_numpy()
    maxv = src.iloc[1].to_numpy()   
    d_range = (maxv - minv)
    df_unnorm = df * d_range
    df_unnorm = df_unnorm + minv
    return df_unnorm
```

```python
test = pd.read_excel(TEST_FILE_ADDRESS, sheet_name="Phase", engine='openpyxl')
test_features = np.expand_dims(test.iloc[:, 1:15].to_numpy(), 2)
test_labels = np.expand_dims(test.iloc[:, 16:30].to_numpy(), 2)

print("test shapes:", test_features.shape, test_labels.shape)
```

```python
model = tf.keras.models.load_model(MODEL_ADDRESS)
model.summary()
```

```python
outcome = model.predict(test_features[:])
print(outcome.shape)
```

```python
outcome =outcome.reshape((2419,14))
re_test_labels = test_labels.reshape((2419,14))
unnormed_outcome = unnormer(outcome, MINMAX_PATH)
unnormed_label = unnormer(re_test_labels, MINMAX_PATH)
err = unnormed_outcome - unnormed_label
mae = np.mean(np.abs(err))
mse = np.mean(np.square(err))

df_clm = ["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10", "P11", "P12", "P13", "P14"]
outcome_df = pd.DataFrame(unnormed_outcome, columns=df_clm)
err_df = pd.DataFrame(err, columns=df_clm)

print("MAE: ", mae)
print("MSE: ", mse)
print("Predicted:\n", outcome_df)
print("Error:\n", err_df)
```
